### head():

- It is a pandas method that returns the first n rows of a DataFrame or Series (default is 5) to quickly preview the data. <br>


```python
# View first few rows
df_adult.head()
```


### sample():

- sample() is a pandas method that returns a random subset of rows from a DataFrame or Series, allowing you to inspect or analyze a representative sample of the data. <br>


```python
# View few rows usingSample
df_adult.sample(5)
```


### info():

- info() is a pandas method that provides a concise summary of a DataFrame, including the number of rows, column names, data types, non-null counts, and memory usage. <br>

```python
# Check column names, data types, and non-null counts
df_adult.info()
```

<br> 

### isnull():

- isnull() is a pandas method that returns a boolean mask indicating whether each value in a DataFrame or Series is missing (NaN or None). <br>

```python
# Check for missing values
df_adult.isnull().sum()
```

<br>

### describe():

- It is a pandas method that generates summary statistics for a DataFrame or Series, such as count, mean, standard deviation, minimum, maximum, and quartiles for numerical data. <br>

```python
# View basic summary statistics
df_adult.describe()
```

<br>
