```python

from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
adult = fetch_ucirepo(id=2) 
  
# data (as pandas dataframes) 
X = adult.data.features 
y = adult.data.targets 
  
# metadata 
print(adult.metadata) 
  
# variable information 
print(adult.variables) 
```

The data is loaded into memory, not as CSV files, but as Pandas DataFrames inside an object(adult).

```python
Understanding adult object
Think of adult as a container with multiple attributes:
adult
├── data
│   ├── features   → Pandas DataFrame (X)
│   ├── targets    → Pandas DataFrame (y)
├── metadata       → Dataset description
├── variables      → Column-level info
```
