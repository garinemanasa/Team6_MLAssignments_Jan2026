“The age column contains 74 unique values, making it difficult to visualize and interpret.
To simplify analysis, I grouped age into meaningful bins such as Young, Early Career, Mid Career, Late Career, and Senior.
This helps reveal patterns in income and workclass across different life stages.”

```python
import pandas as pd

df['age_bin'] = pd.cut(
    df['age'],
    bins=[0, 25, 35, 50, 65, 100],
    labels=['Young (≤25)', 'Early Career (26–35)', 'Mid Career (36–50)',
            'Late Career (51–65)', 'Senior (65+)']
)
```

“The income column contained duplicate categories caused by trailing punctuation, such as <=50K and <=50K..
To ensure consistency, I cleaned the column by stripping whitespace and removing trailing periods.
This standardized the categories and prevented the model from treating them as separate classes.”

```python
df['income'] = (
    df['income']
    .str.strip()
    .str.replace('.', '', regex=False)
    .str.replace(' ', '', regex=False)
    .astype('category')
)
```
